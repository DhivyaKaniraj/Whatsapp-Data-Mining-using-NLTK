# Importing Libraries:

```python
import nltk
import nltk.data
import nltk.tokenize
```

# Loading data:

```python
data=open(r'F:\DHIVYA\DATA SCIENCE\ARTIFICIAL INTELLIGENCE\NLTK\whatsapp data AI.txt',encoding='utf-8')
raw=data.read()
print(raw)
```

# Tokenizing:

```python
from nltk.tokenize import word_tokenize
tokens=word_tokenize(raw)
print(tokens)
```

```python
words=[x.lower() for x in tokens]
print(words)
```

# Cleaning the data:

```python
words=[word for word in words if word.isalpha()]
print(words)
```

```python
import string
useless_words=nltk.corpus.stopwords.words('english') + list(string.punctuation)
print(useless_words)
```

```python
extracted_words=[]
for x in words:
    if x not in useless_words:
        extracted_words.append(x)
    else:
        pass
print(extracted_words)
        
```

```python
len(extracted_words)
```

# Bag of words:

```python
def build_bag_of_words_features_filtered(words):
   return{x:1 for x in words if x not in useless_words}
```

```python
build_bag_of_words_features_filtered(words)
```

# Count on the frequency:

```python
from collections import Counter
word_counter = Counter(extracted_words)
word_counter
```

```python
len(word_counter)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist()
```

```python
for x in extracted_words:
    fdist[x.lower()]+=1
fdist
```

```python
len(fdist)
```

```python
fdist_top10=fdist.most_common(10)
fdist_top10
```
